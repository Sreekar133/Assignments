# 8.Calculate IQR(difference between 75% and 25% quartile)

```python
for feature in features:
  if df[feature].dtypes != "object":
    IQR = df[feature].quantile(0.75) - df[feature].quantile(0.25)
    print(f"IQR for {feature} is {IQR}")
```

# 9.Calculate Standard Deviation for first 4 columns.

```python
fourcolumns = df.iloc[:,0:4]
for feature in fourcolumns:
  std = np.std(df[feature])
  print(f"Standard devidation of {feature} is {std}")
```

# 10.Calculate Variance for first 3 columns.

```python
threecolumns = df.iloc[:,0:3]
for feature in threecolumns:
  var = np.var(df[feature])
  print(f"Variance of {feature} is {var}")
```
